# Visualização de dados - MatPlotLib

```python
 import matplotlib.pyplot as plt
```

## Gŕafico de Linhas + Scatter

```python
x = [1,3,5,7,9]
y = [1,9,25,49,81]

plt.title("Meu primeiro gráfico", color = "r")
plt.xlabel("linha")
plt.ylabel("coluna")
plt.plot(x,y,color = "#4B0082")
plt.scatter(x,y, color = "r", marker = "x",s=70)
plt.show()
```

## Gráfico de Barras

```python
z = [2,4,6,8,10]
w = [4,16,36,64,100]

plt.title("Gráfico de Barras")
plt.xlabel("Linhas")
plt.ylabel("colunas")
plt.bar(z,w)
plt.bar(x,y)
plt.show()
```

## Boxplot:

```python
import random

vector = []

for i in range(10):
    rand = random.randint(0,100000)
    vector.append(rand)

plt.boxplot(vector)
plt.title("Boxplot")
plt.show()
```

## 1- Crescimento da População Brasileira

```python
data = open("Docs/populacao-brasileira.csv").readlines()

x = []
y = []

for i in range(len(data)):
    if i != 0:
        linha = data[i].split(";")
        x.append(int(linha[0]))
        y.append(int(linha[1]))

plt.plot(x,y)
plt.title("Brazilian population growth (1980-2016)")
plt.xlabel("Year")
plt.ylabel("Population  x100.000.000")
plt.savefig("populacaobrasileira.png",dpi=300)
plt.show()
```

## 2- Comparação RNA Humano com RNA de uma Bactéria

```python
entry  = open("Docs/HumanDNA").read()
entry2 = open("Docs/BacteryDNA").read()
out    = open("out.html","w")

cont  = {}
cont2 = {}

for i in ['A','T','C','G']:
    for j in ['A','T','C','G']:
        cont[i+j] = 0

for i in ['A','T','C','G']:
    for j in ['A','T','C','G']:
        cont2[i+j] = 0

entry = entry.replace("\n","")
entry = entry.replace("N","")

entry2 = entry2.replace("\n","")
entry2 = entry2.replace("N","")

for k in range(len(entry)-1):
    cont[entry[k]+entry[k+1]] += 1

for w in range(len(entry2)-1):
    cont2[entry2[w]+entry2[w+1]] += 1
        
plt.bar(cont.keys(),cont.values(),label = "HumanDNA")
plt.bar(cont2.keys(),cont2.values(),label = "BacteryDNA")
plt.legend()

print(cont)
```
