### Importing libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Getting, Knowing and Treating Dataset

```python
sms = pd.read_csv('Docs/spam.csv',encoding = 'latin-1')
sms.head()
```

```python
sms = sms.drop(labels = ['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis = 1)
sms.columns = ['Class','Text']
sms.head()
```

```python
sms.groupby(sms.Class).count()
```

### Tokenizing text messages

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def extractSpam(SpamMessages):
    global spam_words
    words = [word.lower() for word in word_tokenize(SpamMessages) if word.lower() not in stopwords.words('english') and word.lower().isalpha()]
    spam_words = spam_words + words

spam_words = []
sms_spam = sms.Text[sms.Class == 'spam']
sms_spam.apply(extractSpam)
spam_words
```

```python
def extractHam(HamMessages):
    global ham_words
    words = [word.lower() for word in word_tokenize(HamMessages) if word.lower() not in stopwords.words('english') and word.lower().isalpha()]
    ham_words = ham_words + words
    
ham_words = []
sms_ham = sms.Text[sms.Class == 'ham']
sms_ham.apply(extractHam)
ham_words
```

```python
spam_words = np.array(spam_words)
ham_words = np.array(ham_words)

print('Top 10 Spam words:\n', pd.Series(spam_words).value_counts().head(10))
print('\nTop 10 Ham words:\n', pd.Series(ham_words).value_counts().head(10))
```

### Removing punctuation and stopwords & getting key words

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')

def cleanText(message):
    message = message.translate(str.maketrans('','',string.punctuation))
    words = [stemmer.stem(word) for word in message.split() if word.lower() not in stopwords.words('english')]
    return ' '.join(words)

sms.Text = sms.Text.apply(cleanText)
sms.head(10)
```

### Converting text in an representation to ML model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(encoding = 'latin-1', strip_accents = 'unicode', stop_words = 'english')
features = tfidf.fit_transform(sms.Text)
print(features.shape)
```

### Applying the method

```python
#### First we have to convert the classes
def classconverter(message_class):
    if message_class == 'spam':
        return 1
    else:
        return 0

sms.Class = sms.Class.apply(classconverter)
sms.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, sms.Class, stratify = sms.Class, test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB() 
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
