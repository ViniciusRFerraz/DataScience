# Iris Flower Classification
## Unsupervised Learning

### Importing Libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importing Dataset

```python
iris = pd.read_csv('Docs/iris.data', names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Target'])
iris.head()
```

```python
#Giving better names to Target column data
def target(x):
    return x[5].upper() + x[6:].lower()

iris.Target = iris.Target.apply(target)
iris.head()
```

### Viewing Dataset

```python
sns.pairplot(iris, hue='Target')
```

### Making features

```python
X = np.array(iris.drop('Target',axis=1))
```

### Now, we will call K-Means to create and classify our data into classes.
#### As we already know, Iris flower has 3 classes.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state = 0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
iris['K-Classes'] = kmeans.labels_
iris.head()
```

```python
sns.pairplot(iris,hue='K-Classes');
```

```python
kmeans.score(X)
```
