# Iris Flower Classification

### Importing libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Importing Dataset 

```python
iris = pd.read_csv('Docs/iris.data', names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length','Petal_Width','Target'])
iris.head()
```

```python
#creating a function to give better names to the Target column, which will be our classes
def target(x):
    return x[5].upper() + x[6:].lower()

iris.Target = iris.Target.apply(target)
iris.head()
```

```python
from PIL import Image
import requests
from io import BytesIO

response = requests.get('https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Machine+Learning+R/iris-machinelearning.png')
img = Image.open(BytesIO(response.content))
img
```

### Viewing our data

```python
sns.pairplot(iris, hue='Target')
```

### Making Features and Classes

```python
features = np.array(iris[['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']])
classes = np.array(iris.Target)
```

### Classifying with KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(features,classes)
```

```python
#line 0 -> 5.1 | 3.5 | 1.4 | 0.2 | Setosa
#trying with closer numbers
knn.predict([[5.3,3.7,1.3,0.3]])
```

```python
#line 55 -> 5.7 | 2.8 | 4.5 | 1.3 | Versicolor
#closer numbers
knn.predict([[5.6,2.7,4.8,1.1]])
```

```python
#line 123 -> 6.3 | 2.7 | 4.9 | 1.8 | Virginica
knn.predict([[6.2,2.8,4.7,1.6]])
```

### Now, splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, classes, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
knn.fit(X_train,y_train)
```

```python
print('Train -> ',knn.score(X_train,y_train))
print('Test -> ',knn.score(X_test,y_test))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
print('Train -> ',knn.score(X_train,y_train))
print('Test -> ',knn.score(X_test,y_test))
```
